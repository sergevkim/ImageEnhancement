```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split


test_path = '../../Downloads/test.txt'
data_test = pd.read_csv(test_path, sep=',')
data_test.head(10)
```

```python
train_path = '../../Downloads/train.txt'
data = pd.read_csv(train_path, sep=',')
print(set(data['x5']))
data['x5'] = data['x5'].map({'жен.': 1, 'муж.': 0})
data['x11'] = data['x11'].map(
    {'низкий': 1,
     'средний': 2,
     'высокий': 3,
    })

y = data['y']
del data['y']
del data['x6']

X_train, X_val, y_train, y_val = train_test_split(data, y, train_size=0.6, shuffle=True)
X_train.head()
```

```python
X_train = np.asarray(X_train)
X_val = np.asarray(X_val)
y_train = np.asarray(y_train)
y_val = np.asarray(y_val)
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(
    iterations=200,
    learning_rate=0.1,
    depth=4,
    verbose=False)

model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

accuracy_score(y_hat, y_val)
```

```python
roc_auc_score(y_val, y_hat)
```
