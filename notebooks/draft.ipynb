```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from torchvision.transforms import Resize


names = ['9_2000', '99_2000', '825_2000']

for name in names:
    path = f'../data/images/{name}.jpg'
    img = Image.open(path)

    img = Resize(
        size=60)(img)
    
    np_img = np.asarray(img)
    plt.imshow(np_img)
    plt.show()

    print(np_img.shape)
```
